```python
import torchvision
import matplotlib.pyplot as plt

cifar10_dataset = torchvision.datasets.CIFAR10(root='../data', train=True, download=True)
image, label = cifar10_dataset[99]
classes = cifar10_dataset.classes

plt.imshow(image)
plt.title(f'Label: {classes[label]}')
plt.show()
```

```python
import numpy as np

image_np = np.array(image)
patch_size = 8
fig, axes = plt.subplots(4, 4, figsize=(8, 8))

for i in range(4):
    for j in range(4):
        patch = image_np[i*patch_size:(i+1)*patch_size, j*patch_size:(j+1)*patch_size]
        axes[i, j].imshow(patch)
        axes[i, j].axis('off')

plt.show()
```
